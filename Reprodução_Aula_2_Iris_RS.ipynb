## Baixando os pacotes do Scikit-learn

```python
# importação de pacotes
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler #normalização
from sklearn.preprocessing import StandardScaler #padronização
from sklearn.preprocessing import OrdinalEncoder # ordinal encoding
from sklearn.preprocessing import OneHotEncoder # ond-hot enconding e dummy enconding
```

##Limpeza - Tratsametno de Missing

```python
#importante dados da uma url para um dataframe

#url a importar
url_dados = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#label dos atributos do dataset
labels_atributos = ['comprimento_sepala', 'largura_sepala', 'comprimento_petala', 'largura_petala', 'especie' ]

# carga do dataset através da url
iris = pd.read_csv(url_dados, names = labels_atributos)

#exibindo as últimas linhas
iris.tail()

```

```python
#criando um novo dataframe com o iris + uma nova linhas com um valor missing para o próximo exemplo

new_row = {'largura_sepala': 4.0,
          'comprimento_petala': 5.0,
          'largura_petala': 0.4,
          'especie': 'Iris-setosa',
          'comprimento_sepala': None}

df = pd.concat([iris, pd.DataFrame([new_row])], ignore_index = True)

df.tail()
```

```python
#eliminando linhas que tenham algum valor missing

df = df.dropna(how='any')
df.tail()
```

```python
#criando  (novamente) um novo datafram com iris + uma nova linhas com um valor missing para o príoximo exemplo

new_row = {'largura_sepala': 4.0,
          'comprimento_petala': 5.0,
          'largura_petala': 0.4,
          'especie': 'Iris-setosa',
          'comprimento_sepala': None}

df = pd.concat([iris, pd.DataFrame([new_row])], ignore_index = True)

df.tail()
```

```python
#eliminando linhas que tenham todos os valores missing

df = df.dropna(how = 'all')
df.tail()
```

```python
#definindo valores para o preenchimento de missing

values = {'comprimento_sepala': iris['comprimento_sepala'].median(),
          'largura_sepala': 5.0,
          'comprimento_petala': 4.0,
          'largura_petala': 0.1,
          'especie':'Iris-setosa'}

# fazendo o preencimento
df = df.fillna(value = values)
df.tail()
```

## Transformações  Númericas



```python
# dados que iremos usar nos exemplos
data = np.asarray([[100, 0.001],
                   [8, 0.05],
                   [50, 0.005],
                   [88, 0.07],
                   [4, 0.1]])
print(data)
```

##Normalização
y = (x -min) / (max- min)

```python
# Normalização

# definindo o transformador como mim max escaler

scaler = MinMaxScaler()

#transformando  os dados
scaled = scaler.fit_transform(data)
print(scaled)
```

##Padronização
média = soma (x) / contagem (x)

desvio_ padrão = sqrt (soma ((x - média) ^ 2) / contagem (x))

y = (x - média) / desvio padrão


```python
#Padronização

#definindo o transformador como standard scaler
scaler = StandardScaler()

#transformando os dados
scaled = scaler.fit_transform(data)
print(scaled)
```

##Tranformações Categóricas

one hot enconding e dummy enconding

```python
#dados que ieramosuesar nos exmeplos

data = np.asarray([['castanhos'], ['verdes'], ['azuis'], ['azuis'], ['castanhos'], ['azuis']])
print(data)
```

## One hot enconding

```python
#definidndo o transformador com one hot enconding
encoder = OneHotEncoder(sparse = False)

#transformando os dados

onehot = encoder.fit_transform(data)
print(onehot)
```

## Dummy Variable Encoding

```python
#definindo o transformado como one hot encoding (com Dummy variable encoder)
encoder = OneHotEncoder(drop= 'first', sparse = False)

#transformamdo os dados
dummy = encoder.fit_transform(data)
print(dummy)
```
