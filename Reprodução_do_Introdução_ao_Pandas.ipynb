```python
#importação no pandas
import pandas as pd
```

```python
#importando dados uma url pzra um dataframe

#url a importar
url_dados = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#labels dos atributos do dataset
labels_atributos = ['comprimento_sepala', 'laragura_sepala', 'comprimento_petala', 'largura_petala', 'especie']

#carga do data set através da url - há diversos parâmetros no read_csv que podem
#ser interessantes, como sep, usesol, headers
iris = pd.read_csv(url_dados, names = labels_atributos)

```

## Matplotlib

```python
#importação do pyplot
import matplotlib.pyplot as plt

# plotando o gráfico de comprimento x largura da sépala, nome do gráfico é scatter
plt.scatter(iris['comprimento_sepala'], iris['laragura_sepala'])

# incluindo título do gráfico  e rótuços dos eixos
plt.title('Iris: comprimento x largura da sépala')
plt.xlabel('comprimento_sepala')
plt.ylabel('largura_sepala');
```

```python
#criando o dicionário de cores

cores = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

#plotanto cada um dos pontos de dados
for i in range(len(iris['comprimento_sepala'])):
    plt.scatter(iris['comprimento_sepala'][i], iris['laragura_sepala'][i], color = cores[iris['especie'][i]])

# incluindo título do gráfico e rótullos dos eixos
plt.title('Iris: comprimento x largura da sépala separado por classe')
plt.xlabel('comprimento_sepala')
plt.ylabel('laragura_sepala');
```

Trocando as cores por marcadores.

```python
#criando o dicionário de marcadores

marcadores = {'Iris-setosa':'.', 'Iris-versicolor':'+', 'Iris-virginica':'x'}

#plotanto cada um dos pontos de dados
for i in range(len(iris['comprimento_sepala'])):
    plt.scatter(iris['comprimento_sepala'][i], iris['laragura_sepala'][i], color = 'blue', marker=marcadores[iris['especie'][i]])

# incluindo título do gráfico e rótullos dos eixos
plt.title('Iris: comprimento x largura da sépala separado por classe')
plt.xlabel('comprimento_sepala')
plt.ylabel('laragura_sepala');
```

# Gráfico de Linha para evolução histórica

```python
#Gráfico de linhas

#selecionando apenas os atributos  para serem plotados em um novo dataframe, sem a classe
atributos = iris.columns.drop(['especie'])

#criando o eixo x - vai de 0 até a quantidad de linhas do iris - 0 de 149
x_data  = range(0, iris.shape[0])

#plotando cada atibuto

for atributo in atributos:
    plt.plot(x_data, iris[atributo], label = atributo)

 #incluindo título do grpafico e legenda
plt.title('Variação dos atributos dos dataset Iris')
plt.legend();
```

# Histograma  exibe a distribuição de frequencias

```python
# Histograma

# plotando o histograma
plt.hist(iris['comprimento_sepala'], bins = 7, edgecolor = 'black', color = 'yellow')

#incluindo título do gráfico e legenda

plt.title('Distriuição de comprimento da sépala')
plt.ylabel('Frequencia')
plt.xlabel('comprimento_sepala');
```

## Gráfico de Barras

```python
#Atribuindo os valores de x e y
profissoes = ['Engenheiro', 'Médico', 'Professor', 'Vendedor', 'Administrador', 'Outros']
quantidade = [15,30,25,28,33,39]

#Gráfico de Barras

#Crianddo o gráfico de colunas
plt.bar(profissoes, quantidade)

#incluindo título do gráfico e legenda
plt.title("Profissão dos respondentes")
plt.xlabel("Profissao")
plt.ylabel("Númereo de respondentes");
```

```python
# Gráficos de barras horizontal

#criando o gráfico de barras
plt.barh(profissoes, quantidade)

#incluindo título dos gráfico e legenda

plt.title('Profissões dos repondentes')
plt.xlabel("Número de respondentes")
plt.ylabel('Profissão');


```

## Gráfico de Pizza

```python
#criando um gráfico de pizza

plt.pie(quantidade, labels = profissoes);
```

## Em um única figura vários gráficos

```python
#configurando um grupo de subplts em um grid de 1 linha e 2 colunas
#e também o tamanho da figura

fig, ax = plt.subplots(2,2, figsize = (8, 8))

#subplot1: hstograma do comprimento da sepala
ax[0,0].hist(iris['comprimento_sepala'], bins=7, edgecolor = 'black')
ax[0,0].set_title('comprimento_sepala')

#subplot2: histograma da largura_sepala
ax[0,1].hist(iris['laragura_sepala'], bins = 7, edgecolor = 'black')
ax[1,0].set_title('laragura_sepala')

#subplot3: histograma do comprimento_petala
ax[1,0].hist(iris['comprimento_petala'], bins = 7, edgecolor = 'black')
ax[1,0].set_title('comprimento_petala')

#subplot4: histograma do ax[1,1].hist(iris['comprimento_petala'], bins = 7, edgecolor = 'black')
ax[1,1].hist(iris['largura_petala'], bins = 7, edgecolor = 'black');
ax[1,1].set_title('largura_petala');
```

```python
#configurando um grupo de subplts em um grid de 1 linha e 2 colunas
#e também o tamanho da figura  em BOXPLOT

fig, ax = plt.subplots(2,2, figsize = (8, 8))

#subplot1: hstograma do comprimento da sepala
ax[0,0].boxplot(iris['comprimento_sepala'])
ax[0,0].set_title('comprimento_sepala')

#subplot2: histograma da largura_sepala
ax[0,1].boxplot(iris['laragura_sepala'])
ax[1,0].set_title('laragura_sepala')

#subplot3: histograma do comprimento_petala
ax[1,0].boxplot(iris['comprimento_petala'])
ax[1,0].set_title('comprimento_petala')

#subplot4: histograma do ax[1,1].hist(iris['comprimento_petala'], bins = 7, edgecolor = 'black')
ax[1,1].boxplot(iris['largura_petala']);
ax[1,1].set_title('largura_petala');
```
